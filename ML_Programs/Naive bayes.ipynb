```python
import numpy as np
import pandas as pd
```

```python
tennis=pd.read_csv('tennis.csv')
tennis
```

```python
ct_outlook=pd.crosstab(tennis['Outlook'],tennis['play'],margins=True)
ct_outlook
```

```python
ct_temp=pd.crosstab(tennis['Temp'],tennis['play'],margins=True)
ct_temp
```

```python
ct_humidity=pd.crosstab(tennis['Humidity'],tennis['play'],margins=True)
ct_humidity
```

```python
ct_wind=pd.crosstab(tennis['wind'],tennis['play'],margins=True)
ct_wind
```

```python
ct_play=pd.crosstab(index=tennis['play'],columns="count",margins=True)
ct_play
```

```python
p_yes=ct_play.iloc[1,0]/ct_play.iloc[2,0]*ct_outlook.iloc[2,1]/ct_outlook.iloc[3,1]*ct_temp.iloc[0,1]/ct_temp.iloc[3,1]*ct_humidity.iloc[0,1]/ct_humidity[2,1]*ct_wind.iloc[0,1]/ct_wind.iloc[2,1]

print("The probability of tennis is",p_yes)
p_no=ct_play.iloc[0,0]/ct_play.iloc[2,0]*ct_outlook.iloc[2,0]/ct_outlook.iloc[3,0]*ct_temp.iloc[0,0]/ct_temp.iloc[3,0]*ct_humidity.iloc[0,0]/ct_humidity.iloc[2,0]*ct_wind.iloc[0,0]/ct_wind.iloc[2,0]
print('The probability of the tennis that cannot be played',p_no)
if(p_yes>p_no):
    print("The probabilty that the tennis will be played in the following day is high")
else:
    print("The probability that the tennis will not be played in the following day is very low")
```
